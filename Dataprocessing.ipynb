```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data2308 = pd.read_csv('서울특별시 공공자전거 이용정보(일별)_2308.csv', encoding='cp949')
data2308.head()
```

```python
data2308.columns
```

```python
data2308['대여구분코드'].value_counts()
```

```python
data2308 = data2308[data2308['대여구분코드'] != "단체권"]
data2308['대여구분코드'].value_counts()
```

```python
new_order = ['대여일자', '대여소번호', '대여소', '성별', '연령대', ' 이용건수', '운동량', '탄소량','이동거리(M)', '이용시간(분)', '대여구분코드']
df2308 = data2308[new_order]
df2308
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

train_df2308, test_df2308 = train_test_split(df2308, test_size=1/7, random_state=42)

# 결과 출력
print("Train Dataset:")
print(train_df2308)
print("\nTest Dataset:")
print(test_df2308)
```

```python
train_X = train_df2308.iloc[:,:-1]
train_Y = train_df2308.iloc[:,-1]
test_X = test_df2308.iloc[:,:-1]
test_Y = test_df2308.iloc[:,-1]
```

```python
# 트레인 데이터셋 저장
train_X.to_csv('data/S01_train_X.csv', index=False)
train_Y.to_csv('data/S01_train_Y.csv', index=False)
# 테스트 데이터셋 저장
test_X.to_csv('data/S01_test_X.csv', index=False)
test_Y.to_csv('data/S01_test_Y.csv', index=False)
```
